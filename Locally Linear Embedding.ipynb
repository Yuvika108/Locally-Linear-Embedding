```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding

n_samples = 1000
n_neighbors = 10
X, _ = make_swiss_roll(n_samples=n_samples)

lle = LocallyLinearEmbedding(n_neighbors=n_neighbors, n_components=2)
X_r = lle.fit_transform(X)

plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=X[:, 2], cmap=plt.cm.Spectral)
plt.title("Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.subplot(122)
plt.scatter(X_r[:, 0], X_r[:, 1], c=X[:, 2], cmap=plt.cm.Spectral)
#cmap=plt.cm.Spectral
plt.title("Reduced Data (LLE)")
plt.xlabel("Component 1")
plt.ylabel("Component 2")

plt.tight_layout()
plt.show()
```
